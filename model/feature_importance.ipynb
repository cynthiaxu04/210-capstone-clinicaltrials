```python
import joblib
import pandas as pd
import dataframe_image as dfi
import os
```

```python
# Step 1: Load the trained Random Forest model (if it was saved)
model_path = 'model_rf.pkl'
rf_model = joblib.load(model_path)

# Step 2: Access feature importances
feature_importances = rf_model.feature_importances_

# Step 3: Map feature importances to feature names
feature_names_file = 'model_columns_rf.txt'
with open(feature_names_file, 'r') as file:
    feature_names = [line.strip() for line in file]

feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': feature_importances})

# Sort the features by importance (optional)
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

# Print or plot the feature importances
feature_importance_df
```

```python
styled_df = feature_importance_df.style.background_gradient(cmap='Reds') \
                 .set_caption('Feature Importance')

styled_df
```
